```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale

fromage.shape

#fromage1 = StandardScaler().fit_transform(fromage.iloc[:,1:28])

#Center to the mean and component wise scale to unit variance.

#This means that mean value along the axis is subtracted from X and the resulting value is divided by std along the axis.

fromage1 = sklearn.preprocessing.scale(fromage.iloc[:,1:28], axis=0, with_mean=True, with_std=True, copy=True)
```

```python
from matplotlib import pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage


Z = linkage(fromage1,method='ward',metric='euclidean')
Z
```

```python
import scipy.cluster.hierarchy

x = fromage.iloc[:,0].values
    
scipy.cluster.hierarchy.dendrogram(Z, p=30, labels=x, orientation='right', color_threshold=5) #nombre de clusters à determiner par rapport faible augmentation de distance et grosse reduction de groupes. 
# Selon cette explication 4 groupes possibles : Le 1er groupe est constitué de fromages frais. Le 2nd de fromages à pâte molle. Le 3ème de fromages  « durs ». Le 4ème est un peu fourre-tout (de mon point de vue).
#But final donner du sens marketing. 
```

```python
from scipy.cluster.hierarchy import ward, fcluster
import numpy as np
groups = scipy.cluster.hierarchy.fcluster(Z,t=5, criterion='distance') #t = distance threshold. La Concatenation se fait pour donner du sens marketing. 

groups
```

```python
index_of_groups = np.argsort(groups) #Returns the indices that would sort an array.

indexed_groups = pd.DataFrame(groups, fromage.index[index_of_groups]) #D'abord data puis index

indexed_groups

fromage2 = fromage.merge(indexed_groups, how='inner', left_index=True, right_index=True)

fromage2.rename(columns={0: "clusters"}, inplace=True)

fromage2.sort_values(by = 'clusters')

```

```python
from sklearn import cluster 
kmeans= cluster.KMeans(n_clusters=4)
kmeans.fit(fromage1)

# K-means : fonctionne exactement comme les centres mobiles (méthode non hiérarchique) mais recalcule le centre de gravité à chaque nouvel individu.
```

```python
from sklearn import metrics

#utilisation de la métrique "silhouette"
#faire varier le nombre de clusters de 2 à 10

res= np.arange(9,dtype="double")
for k in np.arange(9):
    km = cluster.KMeans(n_clusters=k+2)
    km.fit(fromage1)
    res[k] = metrics.silhouette_score(fromage1,km.labels_) #km.labels recupere la clusterisation des données entrainées par fit
    #The Silhouette Coefficient is calculated using the mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. The Silhouette Coefficient for a sample is (b - a) / max(a, b)
print(res)

```

```python
#graphique
import matplotlib.pyplot as plt
plt.title("Silhouette")
plt.xlabel("# of clusters")
plt.plot(np.arange(2,11,1),res)
plt.show()  #La partition en k = 2 groupes semble la meilleure au sens de la métrique «silhouette». 
            
```
